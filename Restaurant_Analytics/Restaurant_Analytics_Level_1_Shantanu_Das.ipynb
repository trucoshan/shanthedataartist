# Initial Cleaning and Transformation of Data

## Handling Nulls

```python
# Checking the restaurant names and location information of restaurants that have null cuisines
df[df['Cuisines'].isnull()][['Restaurant Name','City','Address','Cuisines']]
```

```python
null_restaurant_cuisines=df[df['Cuisines'].isnull()]['Restaurant Name'].tolist() # List containing names of the restaurants

# Upon research, the restaurants mentioned above were found to have the following Cuisines respectively
null_cuisines=['American, Cafe','American','American, Fast Food','American','American','American, Bar','American','American, Fast food, Portuguese, Hawaiian','American, Mexican']

# Assigning the above values to the respective restaurants
for i in range(0,9):
  mask = (df['Cuisines'].isnull()) & (df['Restaurant Name']==null_restaurant_cuisines[i])
  df.loc[mask, 'Cuisines'] = null_cuisines[i]
```

```python
# Making sure there are no more nulls
df.isnull().sum()
```

## Normalizing Average Cost for Two to USD for easy comparison later on if needed

```python
# Creating a dictionary to help map correct conversion rate to the respective column

conversion_to_usd = {
    'Philippine Peso (PHP, )': 0.01818,
    'Botswana Pula (BWP, P)': 0.0749,
    'Brazilian Real (BRL, R$)': 0.1845,
    'US Dollar (USD, $)': 1.0,
    'Australian Dollar (AUD, A$)': 0.6612,
    'Canadian Dollar (CAD, C$)': 0.7242,
    'Singapore Dollar (SGD, S$)': 0.7806,
    'Indian Rupee (INR, )': 0.01134,
    'Indonesian Rupiah (IDR, Rp)': 0.0000609,
    'New Zealand Dollar (NZD, NZ$)': 0.6944,
    'British Pound Sterling (GBP, )': 1.3567,
    'Qatari Riyal (QAR, QR)': 0.2747,
    'United Arab Emirates Dirham (AED)': 0.2723,
    'South African Rand (ZAR, R)': 0.1432,
    'Sri Lankan Rupee (LKR)': 0.01336,
    'Turkish Lira (TRY, )': 0.2364
}

# Creating a new column and mapping the conversion rates to the respective currencies
df['Conversion_Rate_To_USD'] = df['Currency'].map(conversion_to_usd)

# Creating a new column that translates the Average Cost for Two to one single currency across the dataset
df['Equivalent_In_USD'] = df['Average_Cost_For_Two'] * df['Conversion_Rate_To_USD']
```

# Tasks

## Level 1 - Task 1

**Task: Top Cuisines**
* Determine the top three most common cuisines in the dataset.
* Calculate the percentage of restaurants that serve each of the top cuisines

### Determine the top three most common cuisines in the dataset.

```python
# Taking only cuisines that are strings to avoid complications
df_task1=df[df['Cuisines'].apply(lambda x: isinstance(x,str))]

# Converting the cuisines to list as some restaurants have more than one
df_task1['Cuisines']=df_task1['Cuisines'].apply(lambda x: x.split(', '))

# Exploding the dataset on the basis of the list so each is singled out
df_task1=df_task1.explode('Cuisines')
```

```python
# Showing the top 3 cuisines
df_task1['Cuisines'].value_counts().reset_index().head(3)
```

Therefore, the top 3 cuisines are:

* North Indian
* Chinese
* Fast Food

### Calculate the percentage of restaurants that serve each of the top cuisines

```python
# Calculating the percentage of restaurants that serve the top 3 cuisines
print(f"Percentage of restaurants serving North Indian: {round((df_task1[(df_task1['Cuisines']=='North Indian')]['Restaurant_ID'].nunique()*100)/df_task1['Restaurant_ID'].nunique(),2)} %")
print(f"Percentage of restaurants serving Chinese: {round((df_task1[(df_task1['Cuisines']=='Chinese')]['Restaurant_ID'].nunique()*100)/df_task1['Restaurant_ID'].nunique(),2)} %")
print(f"Percentage of restaurants serving Fast Food: {round((df_task1[(df_task1['Cuisines']=='Fast Food')]['Restaurant_ID'].nunique()*100)/df_task1['Restaurant_ID'].nunique(),2)} %")
```

Therefore:

* Percentage of restaurants serving **North Indian: 41.46 %**
* Percentage of restaurants serving **Chinese: 28.61 %**
* Percentage of restaurants serving **Fast Food: 20.8 %**

### Plot

```python
# Defining the list of top cuisines
top_cuisines = ['North Indian','Chinese','Fast Food']

# Making an empty dictionary for cuisine percentages
cuisine_percentages = {}
# Total number of restaurants
total_restaurants = df_task1['Restaurant_ID'].nunique()

# Assigning respective percentages to the top cuisines
for cuisine in top_cuisines:
    count = df_task1[df_task1['Cuisines'].str.contains(cuisine, na=False)]['Restaurant_ID'].nunique()
    cuisine_percentages[cuisine] = count

# Grouping all other restaurants in 'Others' category
count_others = total_restaurants - sum(cuisine_percentages.values())
cuisine_percentages['Others'] = count_others

# Plotting a pie chart for visualization
plt.figure(figsize=(9,9))
plt.pie(cuisine_percentages.values(),labels=cuisine_percentages.keys(),autopct='%1.2f%%',startangle=90,counterclock=False)
plt.title('Percentage of Restaurants serving Top Cuisines',fontdict={'fontsize':18,'fontweight':'bold'})
plt.legend()
plt.tight_layout()
plt.show()
```

### Calculate the average rating for restaurants in each city.

```python
old_setting = pd.get_option("display.max_rows") # Saving current display setting

pd.set_option("display.max_rows", None) # Showing all rows

# Calculating the weighted average rating for restaurants in each city
# Taking only entries with Votes more than 30 according to CLT
# Not considering entries that have not been rated
df_city_rating=df[(df['Votes']>30)&(df['Rating_Color']!='Not rated')].groupby('City').agg({'WR_Numerator':'sum','Votes':'sum'}).reset_index()
df_city_rating['Weighted_Rating'] = round(df_city_rating['WR_Numerator']/df_city_rating['Votes'],1)
df_city_rating[['City','Weighted_Rating']].sort_values(by='Weighted_Rating',ascending=False)
```

```python
print(f"Average rating of Restaurants with Online Delivery: {float(df1[(df1['Has_Online_Delivery']=='Yes')]['Weighted_Rating'])}")
print(f"Average rating of Restaurants without Online Delivery: {float(df1[(df1['Has_Online_Delivery']=='No')]['Weighted_Rating'])}")
```
